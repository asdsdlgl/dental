```python
def check_shape(sub_graph):
    category0 = [1,2,8,9,10,11,15,16]
    category1 = [3,4,5,6,7,12,13,14]
    for index,row in sub_graph.iterrows():
        if row['category'] == 0:
            if row['class'] in category1:
                if (row['class'] == 3) | (row['class'] == 4):
                    sub_graph.at[index,'class'] = 2
                elif (row['class'] == 5) | (row['class'] == 6) | (row['class'] == 7):
                    sub_graph.at[index,'class'] = 8
                elif (row['class'] == 12) | (row['class'] == 13):
                    sub_graph.at[index,'class'] = 11
                else:
                    sub_graph.at[index,'class'] = 15
        else:
            if row['class'] in category0:
                if (row['class'] == 1) | (row['class'] == 2):
                    sub_graph.at[index,'class'] = 3
                elif (row['class'] == 8) | (row['class'] == 9):
                    sub_graph.at[index,'class'] = 7
                elif (row['class'] == 10) | (row['class'] == 11):
                    sub_graph.at[index,'class'] = 12
                else:
                    sub_graph.at[index,'class'] = 14
    return sub_graph
```

```python
def duplicate(sub_graph, check, table, find_class_table):
    
    distance_df = pd.DataFrame(find_class_table,index=sub_graph.index)
    argdistance_df = pd.DataFrame(np.argsort(find_class_table),index=sub_graph.index)
    #sort_df = pd.concat([sort_df, distance_df], axis=1)
    
    sort_dfRow = sub_graph[sub_graph.duplicated(['class'],keep=False)]
    
    if not sort_dfRow.empty:
        for exist in sub_graph['class']:
            #if exist not in sort_dfRow['class'].unique():
            check[exist-1] = True

        for class_num in sort_dfRow['class'].unique():
            if len(sub_graph.loc[(sub_graph['class']==class_num)])==2:
                if (sub_graph.loc[(sub_graph['class']==class_num)].iloc[0]['min']==sub_graph.loc[(sub_graph['class']==class_num)].iloc[1]['min']) and (sub_graph.loc[(sub_graph['class']==class_num)].iloc[0]['max']==sub_graph.loc[(sub_graph['class']==class_num)].iloc[1]['max']) :
                    continue
                if sub_graph[(sub_graph['class']==class_num)]['category'].iloc[0] == 0:
                    if (class_num == 1) | (class_num == 8) | (class_num == 10) | (class_num == 15):
                        if check[class_num] == False:
                            sub_graph.loc[sub_graph['class'] == class_num, 'class'] = [class_num, class_num+1]
                    elif (class_num == 2) | (class_num == 9) | (class_num == 11) | (class_num == 16):
                        if check[class_num-2] == False:
                            sub_graph.loc[sub_graph['class'] == class_num, 'class'] = [class_num-1, class_num]
                else:
                    if (check[argdistance_df.loc[sub_graph.loc[sub_graph['class']==class_num].index[0],1]] == False) and (argdistance_df.loc[sub_graph.loc[sub_graph['class']==class_num].index[0],1] == class_num-2):
                        sub_graph.loc[sub_graph['class'] == class_num, 'class'] = [class_num-1, class_num]
                    elif (check[argdistance_df.loc[sub_graph.loc[sub_graph['class']==class_num].index[1],1]] == False) and (argdistance_df.loc[sub_graph.loc[sub_graph['class']==class_num].index[1],1] == class_num):
                        sub_graph.loc[sub_graph['class'] == class_num, 'class'] = [class_num, class_num+1]
                    elif (check[argdistance_df.loc[sub_graph.loc[sub_graph['class']==class_num].index[0],2]] == False) and (argdistance_df.loc[sub_graph.loc[sub_graph['class']==class_num].index[0],2] == class_num-2):
                        sub_graph.loc[sub_graph['class'] == class_num, 'class'] = [class_num-1, class_num]
                    elif (check[argdistance_df.loc[sub_graph.loc[sub_graph['class']==class_num].index[1],2]] == False) and (argdistance_df.loc[sub_graph.loc[sub_graph['class']==class_num].index[1],2] == class_num):
                        sub_graph.loc[sub_graph['class'] == class_num, 'class'] = [class_num, class_num+1]
            else:
                min_dis = distance_df.loc[sub_graph['class']==class_num,class_num-1].min()
                for index, row in distance_df.loc[(sub_graph['class']==class_num)].iterrows():
                    if row[class_num-1] != min_dis:

                        if distance_df.loc[index,argdistance_df.loc[index,1]] == distance_df.loc[index,argdistance_df.loc[index,2]]:
                            if sub_graph.loc[index]['category'] == 0:
                                if (check[argdistance_df.loc[index,1]] == False) and (argdistance_df.loc[index,1]+1 in category0):
                                    sub_graph.at[index,'class'] = argdistance_df.loc[index,1] + 1
                                elif (check[argdistance_df.loc[index,2]] == False) and (argdistance_df.loc[index,2]+1 in category0):
                                    sub_graph.at[index,'class'] = argdistance_df.loc[index,2] + 1
                            if sub_graph.loc[index]['category'] == 1:
                                if (check[argdistance_df.loc[index,1]] == False) and (argdistance_df.loc[index,1]+1 in category1):
                                    sub_graph.at[index,'class'] = argdistance_df.loc[index,1] + 1
                                elif (check[argdistance_df.loc[index,2]] == False) and (argdistance_df.loc[index,2]+1 in category1):
                                    sub_graph.at[index,'class'] = argdistance_df.loc[index,2] + 1
                        else:
                            if (sub_graph.loc[index]['category'] == 0) and (argdistance_df.loc[index,1] + 1 in category0):
                                if check[argdistance_df.loc[index,1]] == False:
                                    sub_graph.at[index,'class'] = argdistance_df.loc[index,1] + 1
                            elif (sub_graph.loc[index]['category'] == 1) and (argdistance_df.loc[index,1] + 1 in category1):
                                if check[argdistance_df.loc[index,1]] == False:
                                    sub_graph.at[index,'class'] = argdistance_df.loc[index,1] + 1
                            elif (sub_graph.loc[index]['category'] == 0) and (argdistance_df.loc[index,2] + 1 in category0):
                                if check[argdistance_df.loc[index,2]] == False:
                                    sub_graph.at[index,'class'] = argdistance_df.loc[index,2] + 1
                            elif (sub_graph.loc[index]['category'] == 1) and (argdistance_df.loc[index,2] + 1 in category1):
                                if check[argdistance_df.loc[index,2]] == False:
                                    sub_graph.at[index,'class'] = argdistance_df.loc[index,2] + 1
    
    #dup_sub_graph = sub_graph[sub_graph.duplicated(['class'],keep=False)]
#     print(dup_sub_graph[['image','min','max','mean','category','class']])
#     print('===================================================================')
    return sub_graph
```
