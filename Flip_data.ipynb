```python
import os
import cv2
import glob
import shutil
import random
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
import pandas as pd
import PIL
import json
```

```python
train_image_folder = 'Dental_Data/PBL/PBL_addname/10_clahe_20200810_addname_1'
traing_image_list = glob.glob(train_image_folder+'/*.PNG')
traing_image_list = [strn.replace('\\', "/") for strn in traing_image_list]
print(len(traing_image_list))
traing_image_list
```

```python
for image_path in tqdm(traing_image_list):
    if image_path[-5]=='F':
        continue
        
    img = PIL.Image.open(image_path).convert('RGB')
    img = img.transpose(PIL.Image.FLIP_LEFT_RIGHT)
    x = image_path[:-4]+'F'+'.png'
    
    if int(x.split('_')[-2]) < 17:
        truth = str(17-int(x.split('_')[-2]))
    else:
        truth = str(49-int(x.split('_')[-2]))
    
    new_image_path_list = x.split('_')
    new_image_path_list[-2] = truth
    new_image_path = '_'.join(new_image_path_list)
    
    img.save(new_image_path)
```
